### 3. ĐỌC TỆP PDF

````python
from PyPDF2 import PdfReader  # Import thư viện PyPDF2 để đọc file PDF

def getTextPDF(pdfFileName, password=''):
    # Mở file PDF ở chế độ đọc nhị phân ('rb')
    pdf_file = open(pdfFileName, 'rb')
    
    # Tạo đối tượng PdfReader để đọc nội dung file PDF
    read_pdf = PdfReader(pdf_file)
    
    # Kiểm tra xem file có được mã hóa hay không
    if read_pdf.is_encrypted:
        if password:  # Nếu file bị mã hóa và có mật khẩu, tiến hành giải mã
            read_pdf.decrypt(password)
        else:  # Nếu file bị mã hóa nhưng không có mật khẩu, báo lỗi
            raise ValueError("File này được mã hóa nhưng bạn chưa cung cấp mật khẩu!")

    # Lặp qua tất cả các trang của PDF, sử dụng extract_text() để trích xuất nội dung
    text = [page.extract_text() for page in read_pdf.pages if page.extract_text()]
    
    # Kết hợp nội dung từ tất cả các trang thành một chuỗi duy nhất, cách nhau bởi dấu xuống dòng
    return '\n'.join(text)
````

### 4. ĐỌC TỆP WORD

````python
# Reading woed documents in py
import docx  # Thư viện để xử lý file Word (.docx)

def getTextWord(wordFileName):
    """
    Hàm đọc nội dung của một file Word (.docx) và trả về nội dung dưới dạng chuỗi văn bản.
    
    Tham số:
        wordFileName (str): Đường dẫn đến file Word cần đọc.

    Trả về:
        str: Nội dung văn bản từ file Word, với các đoạn văn cách nhau bởi dấu xuống dòng.
    """
    # Mở file Word
    doc = docx.Document(wordFileName)
    
    # Tạo danh sách để lưu nội dung các đoạn văn (paragraphs)
    fullText = []
    
    # Lặp qua từng đoạn văn trong file Word
    for para in doc.paragraphs:
        fullText.append(para.text)  # Thêm nội dung đoạn văn vào danh sách
    
    # Kết hợp các đoạn văn thành một chuỗi duy nhất, ngăn cách bằng ký tự xuống dòng
    return '\n'.join(fullText)
````

````python
docName = r'D:\TaiLieuHocTap\năm ba 2024-2025\học kì 6\NLP\TH\BUOI3\DATA\Sample.docx'
print('Doccument in full: \n', getTextWord(docName))
````

````python
doc = docx.Document(docName)
print('Number of paragraphs:', len(doc.paragraphs))
print('Paragraph 2:', doc.paragraphs[1].text)
print('Paragraph 2 style:', doc.paragraphs[1].style)
````

````python
# In ra nội dung của đoạn văn đầu tiên trong tài liệu Word
print('Paragraph 1:', doc.paragraphs[0].text)

# Đếm và in số lượng "run" trong đoạn văn đầu tiên
# "Run" là các phần văn bản có cùng định dạng trong Word
print('Number of runs in paragraph 1:', len(doc.paragraphs[0].runs))

# Lặp qua từng "run" trong đoạn văn đầu tiên và in ra nội dung của từng "run"
for idx, run in enumerate(doc.paragraphs[0].runs):
    print('Run %s: %s' % (idx, run.text))  # In chỉ mục (idx) và nội dung của từng "run"
````

````python
# Kiểm tra số lượng runs có sẵn
num_runs = len(doc.paragraphs[0].runs)
print('Total runs in paragraph 1:', num_runs)

# Kiểm tra từng chỉ số có hợp lệ không trước khi truy xuất
if num_runs > 5:
    print('is Run 0 underlined:', doc.paragraphs[0].runs[5].underline)
else:
    print('Run 5 không tồn tại!')

if num_runs > 1:
    print('is Run 2 bold:', doc.paragraphs[0].runs[1].bold)
else:
    print('Run 1 không tồn tại!')

if num_runs > 3:
    print('is Run 7 italic:', doc.paragraphs[0].runs[3].italic)
else:
    print('Run 3 không tồn tại!')
````

## **1. Thao tác với chuỗi (String operations)**  

- **Lệnh `strip()` có tác dụng gì trên chuỗi?**  
  → `strip()` loại bỏ khoảng trắng hoặc ký tự đặc biệt ở đầu và cuối chuỗi.  
  ```python
  text = "  Hello, World!  "
  print(text.strip())  # "Hello, World!"
  ```

- **Làm thế nào để tách một câu thành danh sách các từ?**  
  → Dùng phương thức `split()`:  
  ```python
  sentence = "Hello world! This is NLP."
  words = sentence.split()  
  print(words)  # ['Hello', 'world!', 'This', 'is', 'NLP.']
  ```

- **Nếu muốn thay thế dấu phẩy (`,`) trong chuỗi bằng dấu chấm (`.`), bạn sẽ làm thế nào?**  
  → Dùng `replace()`:  
  ```python
  text = "Hello, world, this is NLP."
  new_text = text.replace(",", ".")
  print(new_text)  # "Hello. world. this is NLP."
  ```

## **2. Đọc tệp PDF**  

- **Python có thể dùng thư viện nào để đọc file PDF?**  
  → Các thư viện phổ biến: `PyPDF2`, `pdfminer.six`, `pdfplumber`.  

- **Đoạn code sau giúp đọc nội dung file PDF, nhưng có sai sót gì không?**  
  ```python
  from PyPDF2 import PdfReader
  def getTextPDF(pdfFileName):
      pdf_file = open(pdfFileName, 'rb')
      read_pdf = PdfReader(pdf_file)
      text = [page.extract_text() for page in read_pdf.pages]
      return '\n'.join(text)
  ```
  → Sai sót:
    1. Không đóng file (`pdf_file.close()` cần được gọi).
    2. `extract_text()` có thể trả về `None`, nên cần kiểm tra.  

- **Nếu muốn trích xuất nội dung của trang đầu tiên trong file PDF, bạn sẽ làm gì?**
